## Topics:
1. fillna to fill missing values using different ways
2. interpolate to make a guess on missing values using interpolation
3. dropna to drop rows with missing values


```python
import pandas as pd

df = pd.read_csv("nyc_weather_data.csv")
df
```

```python
type(df.day[0])
```

```python
# convert day column as date type -- using parse_dates

df = pd.read_csv("nyc_weather_data.csv", parse_dates=["day"])
type(df.day[0])
```

## 1. fillna

```python
df.set_index('day', inplace = True)
new_dataframe = df.fillna(0)
new_dataframe
```

```python
# we can pass a dictionary into fillna

new_dataframe = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_dataframe
```

```python
# carry forward the value---- just put the previous value
#ffill = previous values
#bfill = next values
# pandas fillna documentation

new_dataframe = df.fillna(method="ffill")
new_dataframe
```

```python
# limit helps to restrict the copy items
# so 32 copies 1 time

new_dataframe = df.fillna(method="ffill",limit=1)
new_dataframe
```

```python
#fillna--- axis="columns" coppied values form horizontal

new_df = df.fillna(method="bfill", axis="columns")
new_df
```

## 2.interpolate (statistical analysis)

```python
#estimate missing value according to date

new_df = df.interpolate(method="time")
new_df
```

## 3.dropna

```python
# drop all NaN value
new = df.dropna()
new
```

```python
# drop those rows where all columns become contain Nan value
new = df.dropna(how="all")
new
```

```python
# thresh = 2 -> need two valid value othewwise drop the row

new = df.dropna(thresh=2)
new
```

```python
# further complete the missing values using fillna

dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
