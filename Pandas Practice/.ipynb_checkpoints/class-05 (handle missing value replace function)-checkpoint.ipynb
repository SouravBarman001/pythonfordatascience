## Handle Missing Data:replace function

```python
import pandas as pd
import numpy as np

df = pd.read_csv("class5data.csv")
df
```

```python
# replace -99999 data into NaN
# replace can have a dictionary to specific the column

new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature':[-99999,-88888],
    'event':'0'
},np.NaN)

new_df
```

```python
new_df = df.replace({
    -99999:np.NaN,
    'no event':"Sunny"
    
})

new_df
```

```python
# using regx to remove unnecessary word
'''
new_df = df.replace('[A-Za-z]','',regex = True)

'''



new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
    
},'',regex = True)

new_df
```
