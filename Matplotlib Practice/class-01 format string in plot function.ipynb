```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, color = 'blue', marker='D' ,linestyle = '--')
# short form of red,marker and linestyle plt.plot(x,y,'bD--')
plt.show()
```

```python
plt.plot(x,y,'rD--',markersize=10)
```

```python
#alpha controls the transprency,value=(0 to 1)
plt.plot(x,y,'rD--',markersize=10,alpha=0.5)
```
