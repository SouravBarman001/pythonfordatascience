```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
plt.show()
```

```python
#axis label
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

# if we make a legend ,at first we have to set labels

plt.plot(days,max_t,label ="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend()
plt.show()
```

```python
#axis label
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

# if we make a legend ,at first we have to set labels

plt.plot(days,max_t,label ="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

#change the location of the legend 
#by default it set to be best

plt.legend(loc="upper right",shadow=True,fontsize="small")
plt.show()
```
