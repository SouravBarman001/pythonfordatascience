```python
rangeDate = pd.date_range(start="6/1/2017",end="5/21/2018",freq="B")
rangeDate
```

```python
df.set_index(rangeDate,inplace=True)
df.head()
```

```python
%matplotlib inline
df.Close.plot()
```
